```python
list_name = [
        "x11",
        "x12",
        "x21",
        "x22",
        "x23",
        "x24",
        "x31",
        "x32",
        "x41",
        "x42",
        "x43",
        "x51",
        "x52",
        "x61",
        "x71",
        "x72",
    ]
```

```python
def visualize(wallet_addresses):
    ###
    x11 = []
    x12 = []
    x21 = []
    x22 = []
    x23 = []
    x24 = []
    x31 = []
    x32 = []
    x41 = []
    x42 = []
    x43 = []
    x51 = []
    x52 = []
    x61 = []
    x71 = []
    x72 = []

    filter_criteria = {"_id": {"$in": wallet_addresses}}
    objects = wallet_attributes.find(filter_criteria)
    print(f"Số lượng kết quả: {objects.count()}")

    ###
    for obj in objects:
        x1 = obj.get("creditScorex1")
        x2 = obj.get("creditScorex2")
        x3 = obj.get("creditScorex3")
        x4 = obj.get("creditScorex4")
        x5 = obj.get("creditScorex5")
        x6 = obj.get("creditScorex6")
        x7 = obj.get("creditScorex7")
        x11.append(x1[0])
        x12.append(x1[1])
        x21.append(x2[0])
        x22.append(x2[1])
        x23.append(x2[2])
        x24.append(x2[3])
        x31.append(x3[0])
        x32.append(x3[1])
        x41.append(x4[0])
        x42.append(x4[1])
        x43.append(x4[2])
        x51.append(x5[0])
        x52.append(x5[1])
        x61.append(x6[0])
        x71.append(x7[0])
        x72.append(x7[1])
    list_pramater = [
        x11,
        x12,
        x21,
        x22,
        x23,
        x24,
        x31,
        x32,
        x41,
        x42,
        x43,
        x51,
        x52,
        x61,
        x71,
        x72,
    ]
    for x, name in zip(list_pramater,list_name):
        unique_values = np.unique(x)
        if len(unique_values) == 1:
            plt.hist(x, bins=50, color='blue')
            plt.title('Phân bố giá trị của mảng')
            plt.xlabel(f'Giá trị {name}')
            plt.ylabel('Số lượng')
            plt.grid(True)
            plt.locator_params(axis='y', integer=True)  
            plt.show() 
        else:
            plt.hist(x, bins=50, color='blue')
            plt.title('Phân bố giá trị của mảng')
            plt.xlabel(f'Giá trị {name}')
            plt.ylabel('Số lượng')
            plt.grid(True)
            plt.xlim(-10, 900) 
            plt.locator_params(axis='y', integer=True)  
            plt.show()
```
