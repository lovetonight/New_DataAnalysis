```python
from deposit.crawl_deposit_event import crawl_deposit_event
from borrow.crawl_borrow_event import crawl_borrow_event, crawl_borrow_total
from pymongo import MongoClient
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open(
    "D:/Project/Data-Analysis/New_DataAnalysis/data/whale/whale_740.json", "r"
) as f:
    wallet_addresses_740 = json.load(f)
with open(
    "D:/Project/Data-Analysis/New_DataAnalysis/data/whale/whale_800.json", "r"
) as f:
    wallet_addresses_800 = json.load(f)
```

```python
deposit_740 = crawl_deposit_event(wallet_addresses = wallet_addresses_740, detail=False)
borrow_740 = crawl_borrow_total(wallet_addresses= wallet_addresses_740)
```

```python
deposit_800 = crawl_deposit_event(wallet_addresses = wallet_addresses_800, detail=False)
borrow_800 = crawl_borrow_total(wallet_addresses= wallet_addresses_800)
```

```python
list_pramater = [deposit_740, borrow_740, deposit_800, borrow_800]
list_name = ["deposit_740", "borrow_740", "deposit_800", "borrow_800"]

for tmp, name in zip(list_pramater, list_name):
    arr = list(tmp.values())
    
    min_value = np.min(arr)
    max_value = np.max(arr)
    
    plt.hist(arr, bins=100, color="blue")
    plt.title("Phân bố giá trị của mảng")
    plt.xlabel(f"Giá trị deposit của ví {name}")
    plt.ylabel("Số lượng")
    plt.grid(True)
    
    # Hiển thị giá trị nhỏ nhất và lớn nhất trên đồ thị
    plt.text(min_value, plt.ylim()[1]*0.9, f'Min: {min_value:.2e}', color='red')
    plt.text(max_value, plt.ylim()[1]*0.9, f'Max: {max_value:.2e}', color='red')
    
    plt.show()
```
